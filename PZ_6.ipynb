```python
import random
from collections import Counter

menu = []
ords = []

def add_dish():
    n = input("Название: ")
    p = float(input("Цена: "))
    c = input("Категория: ")
    t = int(input("Время приготовления (мин): "))
    menu.append({"n": n, "p": p, "c": c, "t": t})
    print("Добавлено")

def create_order():
    tbl = input("Номер стола: ")
    items = input("Блюда (через запятую): ").split(',')
    items = [i.strip() for i in items if i.strip()]
    ords.append({"tbl": tbl, "items": items, "status": "active"})
    print("Заказ создан")

def calc_cost(order):
    total = 0
    for it in order["items"]:
        for d in menu:
            if d["n"] == it:
                total += d["p"]
                break
    return total

def popular_dishes():
    cnt = Counter()
    for o in ords:
        if o["status"] == "completed":
            cnt.update(o["items"])
    top = cnt.most_common(3)
    print("Популярные блюда:")
    for dish, n in top:
        print(f"{dish}: {n} заказов")

def revenue_by_cat():
    rev = {}
    for o in ords:
        if o["status"] == "completed":
            for it in o["items"]:
                for d in menu:
                    if d["n"] == it:
                        rev[d["c"]] = rev.get(d["c"], 0) + d["p"]
                        break
    print("Выручка по категориям:")
    for cat, s in rev.items():
        print(f"{cat}: {s} руб.")

def main():
    while True:
        print("\n1. Добавить блюдо")
        print("2. Создать заказ")
        print("3. Рассчитать стоимость заказа")
        print("4. Популярные блюда")
        print("5. Выручка по категориям")
        print("0. Выход")
        ch = input("Выбор: ")
        if ch == "1":
            add_dish()
        elif ch == "2":
            create_order()
        elif ch == "3":
            if not ords:
                print("Нет заказов")
                continue
            for i, o in enumerate(ords):
                print(f"{i+1}. Стол {o['tbl']}, статус {o['status']}")
            try:
                idx = int(input("Номер заказа: ")) - 1
                if 0 <= idx < len(ords):
                    print("Стоимость:", calc_cost(ords[idx]))
                else:
                    print("Неверный номер")
            except:
                print("Ошибка ввода")
        elif ch == "4":
            popular_dishes()
        elif ch == "5":
            revenue_by_cat()
        elif ch == "0":
            break
        else:
            print("Неизвестный выбор")

if __name__ == "__main__":
    main()
```
